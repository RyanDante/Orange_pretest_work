```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#importing my csv file
sales = pd.read_csv(
    'data/sales_data.csv',
    parse_dates=['Date'])
```

```python
sales.head()
```

```python
#What's the mean of Order_Quantity?

sales['Order_Quantity'].mean()
```

```python
#A histogram and a box plot with the Order_Quantity data:

sales['Order_Quantity'].plot(kind='hist', bins=30, figsize=(14,6),ec='black')
```

```python
#How many sales per year do we have?

sales['Year'].value_counts()
```

```python
#A pie plot with the previous data:

sales['Year'].value_counts().plot(kind='pie', figsize=(6,6))
```

```python
# Which country has the most sales quantity of sales?

sales['Country'].value_counts().head(1)
```

```python
sales['Country'].value_counts()
```

```python
# A bar plot of the sales per country:

sales['Country'].value_counts().plot(kind='bar', figsize=(14,6))
```
